```python
import numpy as np
import matplotlib.pyplot as plt
from basic.initializer import normal, xavier, he
from basic.function import sigmoid, relu
```

```python
def plot_activation_hist(fun_activation, fun_init):
    num_node = 100
    hidden_layer_size = 5

    x = np.random.randn(1000, 100)
    activations = {}

    for i in range(hidden_layer_size):
        if i != 0:
            x = activations[i - 1]
        w = fun_init(num_node, num_node)
        z = np.dot(x, w)
        activations[i] = fun_activation(z)

    plt.figure(figsize=(20, 4))
    for i, a in activations.items():
        plt.subplot(1, len(activations), i + 1)
        plt.title("layer {}".format(i + 1))
        plt.hist(a.flatten(), 30, range=(0, 1))
    plt.show()
```

```python
plot_activation_hist(sigmoid, lambda *shape: normal(*shape, std=1))
```

```python
plot_activation_hist(sigmoid, lambda *n: normal(*n, std=0.01))
```

```python
plot_activation_hist(sigmoid, xavier)
```

```python
plot_activation_hist(sigmoid, he)
```

```python
plot_activation_hist(relu, lambda *n: normal(*n, std=0.01))
```

```python
plot_activation_hist(relu, xavier)
```

```python
plot_activation_hist(relu, he)
```
