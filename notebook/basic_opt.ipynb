```python
import numpy as np
import matplotlib.pyplot as plt
from basic.opt import gradient_decent
```

```python
def function_2(x):
    return np.sum(x**2, axis=-1)
```

```python
x = np.array([-3.0, 4.0])
gradient_decent(function_2, x, 0.1, 100)
```

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = function_2(np.array(list(zip(X.ravel(), Y.ravel()))))
Z = Z.reshape((30, 30))
plt.figure()
plt.contour(X, Y, Z, cmap='viridis')

num_iter = 100
path = np.empty((num_iter + 1, 2))
path[0, ] = np.array([-3.0, 4.0])
for i in range(num_iter):
    path[i + 1, :] = gradient_decent(function_2, path[i, :], 0.1, 1)
plt.plot(path[:, 0], path[:, 1], marker='o')

plt.show()
```
