```python
import numpy as np
from net.MultiLayerNet import MultiLayerNet
from data.mnist import read_data
```

```python
data_set = read_data(base_dir='../mnist_data', one_hot=True)
```

```python
network = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100], output_size=10)
```

```python
x_batch = data_set.train.images[:3]
t_batch = data_set.train.labels[:3]
```

```python
x_batch.shape
```

```python
gradient_numerical = network.numerical_gradient(x_batch, t_batch)
```

```python
gradient_backprop = network.gradient(x_batch, t_batch)
```

```python
gradient_numerical
```

```python
gradient_backprop
```

```python
for key in gradient_numerical.keys():
    diff = np.average(np.abs(gradient_backprop[key] - gradient_numerical[key]))
    print("{}: {}".format(key, diff))
```
